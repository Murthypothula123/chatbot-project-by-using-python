```python
# Load the intents file
with open('intents.json') as file:
    data = json.load(file)

print(data)
```

```python
nltk.download('punkt')
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# Initialize lists
words = []
classes = []
documents = []
ignore_words = ['?', '!']

# Preprocess the data
for intent in data['intents']:
    for pattern in intent['patterns']:
        # Tokenize each word in the sentence
        word_list = nltk.word_tokenize(pattern)
        words.extend(word_list)
        # Add documents to list
        documents.append((word_list, intent['tag']))
        # Add to classes if the tag is not already there
        if intent['tag'] not in classes:
            classes.append(intent['tag'])

# Lemmatize and remove duplicates
words = [lemmatizer.lemmatize(w.lower()) for w in words if w not in ignore_words]
words = sorted(list(set(words)))

classes = sorted(list(set(classes)))

print(words)
print(classes)
```

```python
# Initialize training data
training = []
output_empty = [0] * len(classes)

# Create the training set, bag of words for each sentence
for doc in documents:
    bag = []
    # Tokenize each word in the sentence
    word_patterns = doc[0]
    word_patterns = [lemmatizer.lemmatize(word.lower()) for word in word_patterns]

    # Create the bag of words array
    for w in words:
        bag.append(1) if w in word_patterns else bag.append(0)

    # Output is '0' for each tag and '1' for current tag
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1

    training.append([bag, output_row])

# Shuffle the training data
random.shuffle(training)

# Separate the features and labels
train_x = [item[0] for item in training]  # Features
train_y = [item[1] for item in training]  # Labels

# Convert lists to NumPy arrays
train_x = np.array(train_x)
train_y = np.array(train_y)

print("Training data successfully created and converted to NumPy arrays.")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Build a Sequential model
model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
hist = model.fit(train_x, train_y, epochs=200, batch_size=5, verbose=1)

print("Model training completed.")
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Define CSS for enhanced styling
css_code = """
<style>
    .chat-container {
        width: 100%;
        max-width: 600px;
        margin: auto;
        border: 1px solid #ddd;
        border-radius: 5px;
        overflow: hidden;
        box-shadow: 0 0 10px rgba(0,0,0,0.1);
        background-color: #f9f9f9;
    }
    .chat-box {
        height: 400px;
        overflow-y: auto;
        padding: 10px;
        background-color: #fff;
        border-radius: 5px;
        border: 1px solid #ddd;
    }
    .user-message, .bot-message {
        padding: 10px;
        border-radius: 5px;
        margin-bottom: 10px;
        max-width: 80%;
        border: 2px solid transparent;
    }
    .user-message {
        background-color: #e1ffc7;
        border-color: #28a745; /* Green border for user messages */
    }
    .bot-message {
        background-color: #ffffff;
        color: #333;
        border-color: #28a745; /* Green border for bot messages */
    }
    .input-box {
        display: flex;
        padding: 10px;
        border-top: 1px solid #ddd;
        background-color: #fff;
    }
    .input-box input {
        flex: 1;
        padding: 10px;
        border: 1px solid #ddd;
        border-radius: 5px;
    }
    .input-box button {
        padding: 10px 20px;
        border: none;
        border-radius: 5px;
        background-color: #4CAF50;
        color: white;
        cursor: pointer;
        margin-left: 10px;
    }
    .input-box button:hover {
        background-color: #45a049;
    }
</style>
"""

# Create input and output widgets
input_box = widgets.Text(
    value='',
    placeholder='Type your message here...',
    description='',
    layout=widgets.Layout(width='80%')
)

send_button = widgets.Button(
    description='Send',
    button_style='success',
    layout=widgets.Layout(width='20%', height='40px')
)

output_area = widgets.Output(layout=widgets.Layout(border='1px solid #ddd', padding='10px', height='300px', overflow_y='auto', background_color='#fff'))

# Initialize chat history
chat_history = ""

# Function to handle button click
def on_button_clicked(b):
    global chat_history
    with output_area:
        clear_output(wait=True)  # Clear the previous output
        user_input = input_box.value  # Get the user's input
        response = chatbot_response(user_input)  # Get the chatbot's response using the function

        # Append new messages to chat history
        chat_history += f"<div class='user-message'>You: {user_input}</div>"
        chat_history += f"<div class='bot-message'>Chatbot: {response}</div>"

        # Display the updated chat history
        display(HTML(f"<div class='chat-box'>{chat_history}</div>"))

        input_box.value = ''  # Clear the input box after sending the message

# Connect the button click to the function
send_button.on_click(on_button_clicked)

# Display the chat interface with enhanced styling
display(HTML(css_code))
display(widgets.VBox([output_area, widgets.HBox([input_box, send_button])]))
```
